```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

### Two stable 
(1, 5/2) and (-1, -5/2)
### One unstable
(0,0)

There appears to be a theorem that requires the system be differentiable 3 times for there to be three fixed points.

```python
# the x axis is the quantity values
# the y axis is the price values
xvals, yvals = np.meshgrid(np.arange(-4, 4.1, 0.1), np.arange(-4, 4.1, 0.1)) 
```

```python
xdot = -5/2*xvals + yvals
ydot = -3/2*xvals + yvals - xvals**3
# plot dynamics of price and quantity movements
plt.streamplot(xvals, yvals, xdot, ydot)
plt.show()
```
